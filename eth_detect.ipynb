model buliding and hyper parameter tuning

```python
oversample = SMOTE()
print(f'Shape of the training before SMOTE: {norm_train_f.shape, y_train.shape}')

x_tr_resample, y_tr_resample = oversample.fit_resample(norm_train_f, y_train)
print(f'Shape of the training after SMOTE: {x_tr_resample.shape, y_tr_resample.shape}')
```

```python
# Target distribution before SMOTE
non_fraud = 0
fraud = 0

for i in y_train:
    if i == 0:
        non_fraud +=1
    else:
        fraud +=1

# Target distribution after SMOTE
no = 0
yes = 1

for j in y_tr_resample:
    if j == 0:
        no +=1
    else:
        yes +=1


print(f'BEFORE OVERSAMPLING \n \tNon-frauds: {non_fraud} \n \tFauds: {fraud}')
print(f'AFTER OVERSAMPLING \n \tNon-frauds: {no} \n \tFauds: {yes}')
```

ROC

```python
def get_roc_curve(y_test, y_prob):
    false_positive_rate, true_positive_rate, threshold = roc_curve(y_test, y_prob)
    roc_auc = auc(false_positive_rate, true_positive_rate)

    plt.figure(figsize = (10,10))
    plt.title('Receiver Operating Characteristic')
    plt.plot(false_positive_rate, true_positive_rate, color = 'red', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1], linestyle = '--')
    plt.axis('tight')
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
```

logistic regression

```python
LR = LogisticRegression(random_state=42)
LR.fit(x_tr_resample, y_tr_resample)

# Transform test features
norm_test_f = norm.transform(X_test)

preds = LR.predict(norm_test_f)
y_pos_prob = LR.predict_proba(norm_test_f)[:,1]
```

```python
print(y_test.shape)
y_test.value_counts()
```

```python
print("Logistic Regression Model:")
print(classification_report(y_test, preds))
print(confusion_matrix(y_test, preds))
cm = confusion_matrix(y_test, preds)
plt.title("Heatmap of LogisticRegression confusion matrix", fontsize = 15)
sns.heatmap(cm,annot=True,cmap=sns.cubehelix_palette(8),fmt="d")
plt.show()
get_roc_curve(y_test,y_pos_prob)
plt.show()
#metrics.plot_confusion_matrix(LR, norm_test_f, y_test)
```

random forest

```python
RF = RandomForestClassifier(random_state=42)
RF.fit(x_tr_resample, y_tr_resample)
preds_RF = RF.predict(norm_test_f)
y_pos_prob = RF.predict_proba(norm_test_f)[:,1]
print(classification_report(y_test, preds_RF))
print(confusion_matrix(y_test, preds_RF))
cm = confusion_matrix(y_test, preds_RF)
plt.title("Heatmap of random forest confusion matrix", fontsize = 15)
sns.heatmap(cm,annot=True,cmap=sns.cubehelix_palette(8),fmt="d")
plt.show()
get_roc_curve(y_test,y_pos_prob)
plt.show()
```

XGB

```python
xgb_c = xgb.XGBClassifier(random_state=42)
xgb_c.fit(x_tr_resample, y_tr_resample)
preds_xgb = xgb_c.predict(norm_test_f)
y_pos_prob = xgb_c.predict_proba(norm_test_f)[:,1]
print(classification_report(y_test, preds_xgb))
print(confusion_matrix(y_test, preds_xgb))
cm = confusion_matrix(y_test, preds_xgb)
plt.title("Heatmap of XGB confusion matrix", fontsize = 15)
sns.heatmap(cm,annot=True,cmap=sns.cubehelix_palette(8),fmt="d")
plt.show()
get_roc_curve(y_test,y_pos_prob)
plt.show()
```

Light Gradient Boosting Machine

```python
# Define model
lgb_c = lgb.LGBMClassifier(random_state=42)

# Train model
lgb_c.fit(x_tr_resample, y_tr_resample)

# Make predictions and get classification report and confusion matrix
preds_lgb = lgb_c.predict(norm_test_f)
y_pos_prob = lgb_c.predict_proba(norm_test_f)[:,1]
print(classification_report(y_test, preds_lgb))
print(confusion_matrix(y_test, preds_lgb))
cm = confusion_matrix(y_test, preds_lgb)
plt.title("Heatmap of LightGBM confusion matrix", fontsize = 15)
sns.heatmap(cm,annot=True,cmap=sns.cubehelix_palette(8),fmt="d")
plt.show()

# Plot ROC curve
get_roc_curve(y_test,y_pos_prob)
plt.show()
```

Hyperparameters tuning for XGB Classifier

```python
params_grid = {'learning_rate':[0.01, 0.1, 0.5],
              'n_estimators':[100,200],
              'subsample':[0.3, 0.5, 0.9],
               'max_depth':[2,3,4],
               'colsample_bytree':[0.3,0.5,0.7]}

grid = GridSearchCV(estimator=xgb_c, param_grid=params_grid, scoring='recall', cv = 10, verbose = 0)
#grid search optimize XGBoost parameter
grid.fit(x_tr_resample, y_tr_resample)
print(f'Best params found for XGBoost are: {grid.best_params_}')
'''colsample_bytree indicates the proportion of randomly selected features per tree during construction, 
learning_rate indicates the learning rate, max_depth indicates the maximum depth of each tree, 
n_estimators indicates the number of trees, 
and subsample indicates the proportion of randomly selected samples per tree during construction
recall reflects the accuracy of the model in identifying positive samples
'''
print(f'Best recall obtained by the best params: {grid.best_score_}')
```

```python
preds_best_xgb = grid.best_estimator_.predict(norm_test_f)
print(classification_report(y_test, preds_best_xgb))
print(confusion_matrix(y_test, preds_best_xgb))
y_pos_prob = xgb_c.predict_proba(norm_test_f)[:,1]
print(classification_report(y_test, preds_best_xgb))
print(confusion_matrix(y_test, preds_best_xgb))
cm = confusion_matrix(y_test, preds_best_xgb)
plt.title("Heatmap of XGB confusion matrix", fontsize = 15)
sns.heatmap(cm,annot=True,cmap=sns.cubehelix_palette(8),fmt="d")
plt.show()
get_roc_curve(y_test,y_pos_prob)
plt.show()
```
